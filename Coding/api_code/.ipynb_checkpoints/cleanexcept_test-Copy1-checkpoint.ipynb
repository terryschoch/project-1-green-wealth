```python
# Read saved data
country_data_df = pd.read_csv("../output_data/countries.csv")

# Display sample data
country_data_df
```

```python
# Define lists to hold cleaned latitudes and longitudes
lats = []
lons = []

# Create for loop to split .csv Geographical Coordinates into latitude and longitude, define variable for each
for row in country_data_df['Geographical Coordinates']:
    coords = row.split(', ')
    lats_cleaned = coords[0]
    longs_cleaned = coords[1]
    
    # Define if statements to remove Directionals and spaces
    if ' N' in lats_cleaned:
        lats_cleaned = lats_cleaned.replace(' N','')
    if ' E' in longs_cleaned:
        longs_cleaned = longs_cleaned.replace(' E','')
    if ' S' in lats_cleaned:
        lats_cleaned = (f'-{lats_cleaned}').replace(' S','')
    if ' W' in longs_cleaned:
        longs_cleaned = (f'-{longs_cleaned}').replace(' W','')
   
    # Replace dividing spaces with decimals
    lats_cleaned=lats_cleaned.replace(' ','.')
    longs_cleaned=longs_cleaned.replace(' ','.')

    # Append cleaned coordinates to lats and lons lists
    lats.append(lats_cleaned)   
    lons.append(longs_cleaned) 
```

```python
# Define list to hold cleaned GDP data
GDP_Adjusted_to_USD=[]

# Create for loop to split .csv GDP data, define list of characters to delete and isolate wanted column
for row in country_data_df['GDP (Adjusted to $US)']:
    GDP=row.split(' (')
    del_char=['$',',']
    GDP_cleaned=GDP[0]

    # Create for loop to replace unwanted characters
    for char in del_char:
        GDP_cleaned=GDP_cleaned.replace(char,'')
    
    # Define if statements to remove unwanted word, convert to float, adjust decimal, round
    if ' billion' in GDP_cleaned:
        GDP_cleaned=round(float(GDP_cleaned.replace(' billion','')),2)
    elif ' million' in GDP_cleaned:
        GDP_cleaned=round(float(GDP_cleaned.replace(' million',''))/1000,2)        
    else:        
        GDP_cleaned=round((float(GDP_cleaned)/1000000000),2)
    
    # Append cleaned gdp to GDP_Adjusted_to_USD list
    GDP_Adjusted_to_USD.append(GDP_cleaned) 
```

```python
Percent_of_Waste_Recycled_Annually=[float(row.split('%')[0]) for row in country_data_df['% of Waste Recycled Annually']]
Particulate_Matter_Emissions=[float(row.split(' micrograms per cubic meter (2016 est.)')[0].replace(',','')) for row in country_data_df['Particulate Matter Emissions']]
C02_Emmissions=[float(row.split(" megatons (2016 est.)")[0].replace(',','')) for row in country_data_df['C02 Emissions']]
Methane_Emissions=[float(row.split(' megatons (2020 est.)')[0].replace(',','')) for row in country_data_df['Methane  Emissions']]
Electricity_by_Fossil_Fules=[float(row.split('%')[0]) for row in country_data_df['Electricity by Fossil Fuels']]
Electricity_by_Nuclear=[float(row.split('%')[0]) for row in country_data_df['Electricity by Nuclear']]
Electricity_by_Solar=[float(row.split('%')[0]) for row in country_data_df['Electricity by Solar']]
Electricity_by_Wind=[float(row.split('%')[0]) for row in country_data_df['Electricity by Wind']]
Electricity_by_Hydroelectricty=[float(row.split('%')[0]) for row in country_data_df['Electricity by Hydroelectricty']]
Electricity_by_TideandWave=[float(row.split('%')[0]) for row in country_data_df['Electricity by Tide and Wave']]
Electricity_by_Geothermal=[float(row.split('%')[0]) for row in country_data_df['Electricity by Geothermal']]
Electricity_by_Biomass_and_Waste=[float(row.split('%')[0]) for row in country_data_df['Electricity by Biomass and Waste']]
Urban_Population=[float(row.split('%')[0]) for row in country_data_df['Urban Population (%)']]
Net_Migration=[float(row.split(' migrant(s)/1,000 population (2022 est.)')[0].replace(',','')) for row in country_data_df['Net Migration']]
```

```python
country_data_cleaned_df=country_data_df.drop(['Geographical Coordinates', 'Unnamed: 0',
       'GDP (Adjusted to $US)', 'Real GDP (Purchasing Power Parity) 2020',
       'Real GDP (Purchasing Power Parity) 2019',
       'Real GDP (Purchasing Power Parity) 2018', 'Real GDP Per Capita 2020',
       'Real GDP Per Capita 2019', 'Real GDP Per Capita 2018',
       'Waste Generated Annually', 'Waste Recycled Annually',
       '% of Waste Recycled Annually', 'Particulate Matter Emissions',
       'C02 Emissions', 'Methane  Emissions',
       'Environmental Agreements (signed but not ratified)',
       'Electricity by Fossil Fuels', 'Electricity by Nuclear',
       'Electricity by Solar', 'Electricity by Wind',
       'Electricity by Hydroelectricty', 'Electricity by Tide and Wave',
       'Electricity by Geothermal', 'Electricity by Biomass and Waste',
       'Urban Population (%)', 'Net Migration'],axis=1)
country_data_cleaned_df['Lat']=lats
country_data_cleaned_df['Long']=lons
country_data_cleaned_df['GDP (billion USD)']=GDP_Adjusted_to_USD
country_data_cleaned_df['Real GDP 2020 (billion USD)']=Real_GDP_2020
country_data_cleaned_df['Real GDP 2019 (billion USD)']=Real_GDP_2019
country_data_cleaned_df['Real GDP 2018 (billion USD)']=Real_GDP_2018
country_data_cleaned_df['Real GDP Per Capita 2020']=GDP_per_Capita_2020
country_data_cleaned_df['Real GDP Per Capita 2019']=GDP_per_Capita_2019
country_data_cleaned_df['Real GDP Per Capita 2018']=GDP_per_Capita_2018
country_data_cleaned_df['Wasted Generated Annually (million tons)']=Waste_generated_annually
country_data_cleaned_df['Wasted Recycled Annually (million tons)']=Waste_recycled_annually
country_data_cleaned_df['% of Wasted Recycled Annually']=Percent_of_Waste_Recycled_Annually
country_data_cleaned_df['Particulate Matter Emissions (micrograms per m3)']=Particulate_Matter_Emissions
country_data_cleaned_df['C02 Emissions (megatons)']=C02_Emmissions
country_data_cleaned_df['Methane  Emissions (megatons)']=Methane_Emissions
country_data_cleaned_df['Electricity by Unclean Energy (%)']=Electricity_by_Fossil_Fules
country_data_cleaned_df['Electricity by Clean Energy (%)']=[100-row for row in Electricity_by_Fossil_Fules]
country_data_cleaned_df['Electricity by Fossil Fules (%)']=Electricity_by_Fossil_Fules
country_data_cleaned_df['Electricity by Nuclear (%)']=Electricity_by_Nuclear
country_data_cleaned_df['Electricity by Solar (%)']=Electricity_by_Solar
country_data_cleaned_df['Electricity by Wind (%)']=Electricity_by_Wind
country_data_cleaned_df['Electricity by Hydroelectricty (%)']=Electricity_by_Hydroelectricty
country_data_cleaned_df['Electricity by Tide and Wave (%)']=Electricity_by_TideandWave
country_data_cleaned_df['Electricity by Geothermal (%)']=Electricity_by_Geothermal
country_data_cleaned_df['Electricity by Biomass and Waste (%)']=Electricity_by_Biomass_and_Waste
country_data_cleaned_df['Urban Population (%)']=Urban_Population
country_data_cleaned_df['Net Migration']=Net_Migration

country_data_cleaned_df.head()
```

```python
import numpy as np
agreement_list=[]
total_agreement=[]
for row in country_data_cleaned_df['Environmental Agreements (party to)']:
    agreement_raw=row.split(',')
    total_agreement.append(len(agreement_raw))
    for agreement_name in agreement_raw:
        agreement_list.append(agreement_name)
country_data_cleaned_df['Total Agreement']=total_agreement

agreement_unique=[]
for item in agreement_list:
    if item not in agreement_unique:
        agreement_unique.append(item)
agreement_unique=sorted(agreement_unique)

for item in agreement_unique:
    country_data_cleaned_df[f'{item}']=np.nan
    for index,row in enumerate(country_data_cleaned_df[f'{item}']):
        if item in country_data_cleaned_df['Environmental Agreements (party to)'][index]:
            country_data_cleaned_df[f'{item}'][index]=1
        else:
            country_data_cleaned_df[f'{item}'][index]=0
```
