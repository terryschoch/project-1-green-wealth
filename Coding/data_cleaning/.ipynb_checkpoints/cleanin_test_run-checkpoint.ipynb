```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read saved data
country_data_df = pd.read_csv("../output_data/countries.csv", index_col="Country_ID")

# Display sample data
country_data_df.head()
```

```python
# Define lists to hold cleaned latitudes and longitudes
lats = []
lons = []

# Create for loop to split .csv Geographical Coordinates into latitude and longitude, define variable for each
for row in country_data_df['Geographical Coordinates']:
    coords = row.split(', ')
    lats_cleaned = coords[0]
    longs_cleaned = coords[1]
    
    # Define if statements to remove Directionals and spaces
    if ' N' in lats_cleaned:
        lats_cleaned = lats_cleaned.replace(' N','')
    if ' E' in longs_cleaned:
        longs_cleaned = (f'-{longs_cleaned}').replace(' E','')
    if ' S' in lats_cleaned:
        lats_cleaned = (f'-{lats_cleaned}').replace(' S','')
    if ' W' in longs_cleaned:
        longs_cleaned = (f'-{longs_cleaned}').replace(' W','')
   
    # Replace dividing spaces with decimals
    lats_cleaned=lats_cleaned.replace(' ','.')
    longs_cleaned=longs_cleaned.replace(' ','.')

    # Append cleaned coordinates to lats and lons lists
    lats.append(lats_cleaned)   
    lons.append(longs_cleaned) 
    
# print(lats, lons)
```

```python
# Define list to hold cleaned GDP data
GDP_Adjusted_to_USD=[]

# Create for loop to split .csv GDP data, define variables for first column and characters to delete
for row in country_data_df['GDP (Adjusted to $US)']:
    GDP=row.split(' (')
    del_char=['$',',']
    GDP_cleaned=GDP[0]

    # Create for loop replace unwanted characters
    for char in del_char:
        GDP_cleaned=GDP_cleaned.replace(char,'')
    
    # Define if statements to remove unwanted word, convert to float, format to desired output and round
    if ' billion' in GDP_cleaned:
        GDP_cleaned=round(float(GDP_cleaned.replace(' billion','')),2)
    elif ' million' in GDP_cleaned:
        GDP_cleaned=round(float(GDP_cleaned.replace(' million',''))/1000,3)        
    else:        
        GDP_cleaned=round((float(GDP_cleaned)/1000000000),2)
    
    # Append cleaned gdp to GDP_Adjusted_to_USD list
    GDP_Adjusted_to_USD.append(GDP_cleaned) 

# print(GDP_Adjusted_to_USD)
```

```python
# # Waste and Recycling Info
# Split .csv Waste Generated Annually data and define variabe to store the first returned column only
wastegen = (country_data_df['Waste Generated Annually'].str.split(' ', expand=True))[0]

# Split .csv Waste Recycled Annually data and define variabe to store the first returned column only
wasterec = (country_data_df['Waste Recycled Annually'].str.split(' ', expand=True))[0]

# Split .csv % of Waste Recycled Annually data and define variabe to store the first returned column only
perrec = (country_data_df['% of Waste Recycled Annually'].str.split(' ', expand=True))[0]
```

```python
# # Emissions Info
# Split .csv Particulate Matter Emissions data and define variabe to store the first returned column only
pmems = (country_data_df['Particulate Matter Emissions'].str.split(' ', expand=True))[0]

# Split .csv Particulate Matter Emissions data and define variabe to store the first returned column only
co2ems = (country_data_df['C02 Emissions'].str.split(' ', expand=True))[0]

# Split .csv Particulate Matter Emissions data and define variabe to store the first returned column only
mems = (country_data_df['Methane  Emissions'].str.split(' ', expand=True))[0]
```

```python
# # Environmental Agreements Info
# Define list to hold total of Agreed to Environmental Agreements data and value of variabe "x"
envagtot = []
x = 0

# Create a while loop to split .csv Agreed to Environmental Agreements data 
# count the total # of agreements per country by combining len and .count function
while x < country_data_df['Country'].count():
    envagnums = len(country_data_df['Environmental Agreements (party to)'].str.split(', ')[x])
    
    # Append results to list
    envagtot.append(envagnums)  
    x += 1


# Define list to hold total of Unratified Environmental Agreements data and value of variabe "y"
envagnrtot = []
y = 0

# Create a while loop to split .csv Unratified Environmental Agreements data; 
# count the total # of agreements per country by combining len and .count function
while y < country_data_df['Country'].count():
    envagnrnums = len(country_data_df['Environmental Agreements (signed but not ratified)'].str.split(', ')[y])
    
    # Append results to list
    envagnrtot.append(envagnrnums)  
    y += 1

# print(envagtot, envagnrtot)
```

```python
# # Electricity Generation Info
# Split .csv Electricity by Fossil Fuels data and define variabe to store the first returned column only
ebff = (country_data_df['Electricity by Fossil Fuels'].str.split(' ', expand=True))[0]

# Split .csv Electricity by Nuclear data and define variabe to store the first returned column only
ebn = (country_data_df['Electricity by Nuclear'].str.split(' ', expand=True))[0]

# Split .csv Electricity by Solar data and define variabe to store the first returned column only
ebs = (country_data_df['Electricity by Solar'].str.split(' ', expand=True))[0]

# Split .csv Electricity by Wind data and define variabe to store the first returned column only
ebw = (country_data_df['Electricity by Wind'].str.split(' ', expand=True))[0]

# Split .csv Electricity by Hydroelectricty data and define variabe to store the first returned column only
ebhe = (country_data_df['Electricity by Hydroelectricty'].str.split(' ', expand=True))[0]

# Split .csv Electricity by Tide and Wave data and define variabe to store the first returned column only
ebtw = (country_data_df['Electricity by Tide and Wave'].str.split(' ', expand=True))[0]

# Split .csv Electricity by Geothermal data and define variabe to store the first returned column only
ebgt = (country_data_df['Electricity by Geothermal'].str.split(' ', expand=True))[0]

# Split .csv Electricity by Biomass and Waste data and define variabe to store the first returned column only
ebbw = (country_data_df['Electricity by Biomass and Waste'].str.split(' ', expand=True))[0]
```

```python
# # Urbanization Info
# Define list to hold cleaned Urban Population data
urb_pop = []

# Create for loop to split .csv Urban Population data, define variables for first column
for row in country_data_df['Urban Population (%)']:
    up = row.split(' ')
    up_cleaned = up[0]
    
    # Create for loop to replace unwanted character
    for row in up_cleaned:
        up_cleaned = up_cleaned.replace('%','')

     # Append cleaned urban population to urb_pop list
    urb_pop.append(up_cleaned)    
    
# print(urb_pop)

# Split .csv Net Migration data and define variabe to store the first returned column onlyupsplit = country_data_df['Urban Population (%)'].str.split(' ', expand=True)
nm = (country_data_df['Net Migration'].str.split(' ', expand=True))[0]
```

```python
cleaned_df = pd.DataFrame({'Country': country_data_df['Country'], 
                        'Lat': lats, 
                        'Lon': lons, 
                        'GDP': GDP_Adjusted_to_USD,
                        'Annual Waste Generated (tons)': wastegen,
                        'Annual Waste Recycled (tons)': wasterec,
                        'Waste Recycled %': perrec,
                        'Particulate Matter Emissions (micrograms per cubic meter)': pmems,
                        'C02 Emissions (megatons)': co2ems,
                        'Methane  Emissions (megatons)': mems,
                        "Environmental Agreements (party to)": envagtot,
                        "Environmental Agreements (signed but not ratified)": envagnrtot,
                        'Electricity by Fossil Fuels': ebff,
                        "Electricity by Nuclear": ebn,
                        "Electricity by Solar": ebs,
                        "Electricity by Wind": ebw,
                        "Electricity by Hydroelectricty": ebhe,
                        "Electricity by Tide and Wave": ebtw,
                        "Electricity by Geothermal": ebgt,
                        "Electricity by Biomass and Waste": ebbw, 
                        "Urban Population (%)": urb_pop,
                        "Net Migration (/1000 Population)": nm
                       })
cleaned_df
```

```python
# Export the cleaned_df into a csv
cleaned_df.to_csv("../output_data/cleaned_countries.csv", index_label="Country_ID")
```
