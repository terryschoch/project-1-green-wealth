```python
# Dependencies and Setup
import pandas as pd
import re
```

```python
# Read saved data
country_data_df = pd.read_csv("../output_data/countries.csv")

# Display sample data
country_data_df
```

## Geo Location Data

```python
# Define lists to hold cleaned latitudes and longitudes
lats = []
lons = []

# Create for loop to split .csv Geographical Coordinates into latitude and longitude, define variable for each
for row in country_data_df['Geographical Coordinates']:
    coords = row.split(', ')
    lats_cleaned = coords[0]
    longs_cleaned = coords[1]
    
    # Define if statements to remove Directionals and spaces
    if ' N' in lats_cleaned:
        lats_cleaned = lats_cleaned.replace(' N','')
    if ' E' in longs_cleaned:
        longs_cleaned = longs_cleaned.replace(' E','')
    if ' S' in lats_cleaned:
        lats_cleaned = (f'-{lats_cleaned}').replace(' S','')
    if ' W' in longs_cleaned:
        longs_cleaned = (f'-{longs_cleaned}').replace(' W','')
   
    # Replace dividing spaces with decimals
    lats_cleaned=lats_cleaned.replace(' ','.')
    longs_cleaned=longs_cleaned.replace(' ','.')

    # Append cleaned coordinates to lats and lons lists
    lats.append(float(lats_cleaned))   
    lons.append(float(longs_cleaned))
```

 ## Economic Data

```python
# Define list to hold cleaned GDP data
GDP_Adjusted_to_USD=[]

# Create for loop to split .csv GDP data, define characters to delete and isolate wanted column
for row in country_data_df['GDP (Adjusted to $US)']:
    GDP=row.split(' (')
    del_char=['$',',']
    GDP_cleaned=GDP[0]

    # Create for loop to replace unwanted characters
    for char in del_char:
        GDP_cleaned=GDP_cleaned.replace(char,'')
    
    # Define if statements to remove unwanted word, convert to float, adjust decimal, round
    if ' billion' in GDP_cleaned:
        GDP_cleaned=round(float(GDP_cleaned.replace(' billion','')),2)
    elif ' million' in GDP_cleaned:
        GDP_cleaned=round(float(GDP_cleaned.replace(' million',''))/1000,2)        
    else:        
        GDP_cleaned=round((float(GDP_cleaned)/1000000000),2)
    
    # Append cleaned gdp to GDP_Adjusted_to_USD list
    GDP_Adjusted_to_USD.append(GDP_cleaned) 
```

```python
# Define list to hold cleaned Real GDP data
Real_GDP=[]

# Create for loop to split .csv Real GDP data, define characters to delete and isolate wanted column
for row in country_data_df['Real GDP (Purchasing Power Parity) 2020']:
    RGDP=re.split("'text': | \(20", row)
    del_char=['$',"'", ',']
    RGDP_cleaned=RGDP[1]
    
    # Create for loop replace unwanted characters
    for char in del_char:
        RGDP_cleaned=RGDP_cleaned.replace(char,'')
    
    # Define if statements to remove remaining unwanted word, convert to float, adjust decimal, round
    if ' billion' in RGDP_cleaned:
        RGDP_cleaned=(RGDP_cleaned.replace(' billion',''))
        RGDP_cleaned=round(float(RGDP_cleaned),2)
    elif ' million' in RGDP_cleaned:
        RGDP_cleaned=(RGDP_cleaned.replace(' million',''))
        RGDP_cleaned=round(float(RGDP_cleaned)/1000,2)                        
    else:        
        RGDP_cleaned=round(float(RGDP_cleaned)/1000000000,2)
    
    # Append RGDP_cleaned to Real_GDP list
    Real_GDP.append(RGDP_cleaned) 
```

```python
# Define list to hold cleaned Real GDP per Capita data
GDP_per_Capita = []

# Create for loop to split .csv Real GDP per Capita data, define characters to delete and isolate wanted column
for row in country_data_df['Real GDP Per Capita 2020']:
    GDPPC = re.split("'text': | \(20", row)
    del_char = ['$',"'", ',']
    GDPPC_cleaned = GDPPC[1]
    
    # Create for loop replace unwanted characters
    for char in del_char:
        GDPPC_cleaned = GDPPC_cleaned.replace(char,'')
        
    # Append GDPPC_cleaned to GDP_per_Capita list
    GDP_per_Capita.append(float(GDPPC_cleaned))
```

## Waste and Recycling Data

```python
# Define list to hold cleaned Waste Generated data
Waste_generated_annually=[]

# Create for loop to split .csv Waste Generated data, isolate wanted column, eliminate commas
for row in country_data_df['Waste Generated Annually']:
    Waste=row.split(' tons')
    Waste_cleaned=Waste[0]
    Waste_cleaned=Waste_cleaned.replace(',','')
    
    # Define if statements to remove remaining unwanted word, convert to float, adjust decimal, round
    if ' million' in Waste_cleaned:
        Waste_cleaned=Waste_cleaned.replace(' million','')
        Waste_cleaned=round(float(Waste_cleaned),2)        
    else:
        Waste_cleaned=float(Waste_cleaned)
        Waste_cleaned=float(round(Waste_cleaned/1000000,2))
        
    # Append Waste_cleaned to Waste_generated_annually list
    Waste_generated_annually.append(Waste_cleaned)

# Add Waste column to original DF with cleaned waste generated data
country_data_df['Waste'] = Waste_generated_annually
```

```python
# Define list to hold cleaned Waste Recycled data
Waste_recycled_annually=[]

# Drop NaN values in column to allow for float conversion
recycled_NaN = country_data_df['Waste Recycled Annually'].fillna('NA tons')
    
# Create for loop to split .csv Waste Recycled data, isolate wanted column, eliminate commas    
for row in recycled_NaN:    
    Recycled=row.split(' tons')            
    Recycled_cleaned=Recycled[0]
    Recycled_cleaned=Recycled_cleaned.replace(',','')
    
    # Define if statements to remove remaining unwanted word, convert to float, adjust decimal, round
    if ' million' in Recycled_cleaned:        
        Recycled_cleaned=Recycled_cleaned.replace(' million','')
        Recycled_cleaned=float(Recycled_cleaned)
        Recycled_cleaned=round(Recycled_cleaned,2)
    elif 'NA' in Recycled_cleaned:
        Recycled_cleaned=Recycled_cleaned.replace('NA','0')
        Recycled_cleaned=float(Recycled_cleaned)
    else:        
        Recycled_cleaned=float(Recycled_cleaned)
        Recycled_cleaned=float(round(Recycled_cleaned/1000000,2))
        
    # Append Waste_cleaned to Waste_recycled_annually list   
    Waste_recycled_annually.append(Recycled_cleaned)

# Add Recycled column to original DF with cleaned waste recycled data
country_data_df['Recycled'] = Waste_recycled_annually
```

```python
# Define list to hold cleaned Percent of Waste Recycled data
Percent_of_Waste_Recycled_Annually=[]

# Drop NaN values in column to allow for float conversion
percent_NaN = country_data_df['% of Waste Recycled Annually'].fillna('NA')
    
# Create for loop to split .csv % Waste Recycled data, isolate wanted column,   
for row in percent_NaN:    
    RecycledPer=row.split('%')            
    RecycledPer_cleaned=RecycledPer[0]
    
    # Define if statements to remove remaining unwanted word, convert to float, adjust decimal, round
    if 'NA' in RecycledPer_cleaned:
        RecycledPer_cleaned=RecycledPer_cleaned.replace('NA','0')     
        
    # Append Waste_cleaned to Waste_recycled_annually list   
    Percent_of_Waste_Recycled_Annually.append(float(RecycledPer_cleaned))
```

```python
# Determine Net Waste produced by .loc of recycling data and simple subtraction
dev_waste_data = country_data_df.loc[country_data_df["Recycled"] > 0, ["Waste", "Recycled"]]
dev_netwaste = dev_waste_data['Waste'] - dev_waste_data['Recycled']
```

## Environmental Agreements Info

```python
# Define list to hold total of Agreed to Environmental Agreements data and value of variabe "x"
envagreementstot = []
x = 0

# Create a while loop to split .csv Agreed to Environmental Agreements data 
# count the total # of agreements per country by combining len and .count function
while x < country_data_df['Country'].count():
    envagreements = len(country_data_df['Environmental Agreements (party to)'].str.split(', ')[x])
    
    # Append results to list
    envagreementstot.append(envagreements)  
    x += 1

# Create column in country_data_df and add list
country_data_df['Environment Agreements #']=envagreementstot


# Define list to hold total of Unratified Environmental Agreements data and value of variabe "y"
envagreementsnrtot = []
y = 0

# Create a while loop to split .csv Unratified Environmental Agreements data; 
# count the total # of agreements per country by combining len and .count function
while y < country_data_df['Country'].count():
    envagreementsnr = len(country_data_df['Environmental Agreements (signed but not ratified)'].str.split(', ')[y])
    
    # Append results to list
    envagreementsnrtot.append(envagreementsnr)  
    y += 1

# Create column in country_data_df and add list
# country_data_df['Environment Agreements Not Ratified #']=envagreementsnrtot
country_data_df['Environment Agreements Not Ratified #']=envagreementsnrtot
    
# Find individual sums of agreement totals
total_agreements = country_data_df['Environment Agreements #'] + country_data_df['Environment Agreements Not Ratified #']
```

## Emissions, Power Generation, Urbanization/Migration Data, Development Status

```python
# Clean remaining columns via list comprehension of float, split, replace functions
Particulate_Matter_Emissions=[float(row.split(' micrograms per cubic meter (2016 est.)')[0].replace(',','')) for row in country_data_df['Particulate Matter Emissions']]
C02_Emmissions=[float(row.split(" megatons (2016 est.)")[0].replace(',','')) for row in country_data_df['C02 Emissions']]
Methane_Emissions=[float(row.split(' megatons (2020 est.)')[0].replace(',','')) for row in country_data_df['Methane Emissions']]
Electricity_by_Fossil_Fules=[float(row.split('%')[0]) for row in country_data_df['Electricity by Fossil Fuels']]
Electricity_by_Nuclear=[float(row.split('%')[0]) for row in country_data_df['Electricity by Nuclear']]
Electricity_by_Solar=[float(row.split('%')[0]) for row in country_data_df['Electricity by Solar']]
Electricity_by_Wind=[float(row.split('%')[0]) for row in country_data_df['Electricity by Wind']]
Electricity_by_Hydroelectricty=[float(row.split('%')[0]) for row in country_data_df['Electricity by Hydroelectricty']]
Electricity_by_TideandWave=[float(row.split('%')[0]) for row in country_data_df['Electricity by Tide and Wave']]
Electricity_by_Geothermal=[float(row.split('%')[0]) for row in country_data_df['Electricity by Geothermal']]
Electricity_by_Biomass_and_Waste=[float(row.split('%')[0]) for row in country_data_df['Electricity by Biomass and Waste']]
Urban_Population=[float(row.split('%')[0]) for row in country_data_df['Urban Population (%)']]
Net_Migration=[float(row.split(' migrant(s)/1,000 population (2022 est.)')[0].replace(',','')) for row in country_data_df['Net Migration']]
```

```python
# Define list to hold development status of countries
country_status = []
 
# Create for loop to define status based on Real GDP per Capita column 
for row in GDP_per_Capita:       
    # Define if statements to categorize countries
    if row <= 4000:
        country_status.append('LDC')
    elif row >= 36500:
        country_status.append('Developed')
    else:
        country_status.append("Developing")
```

## Final DataFrame formation

```python
# Define final_contry_data DataFrame with dropped dirty columns
final_country_data_df=country_data_df.drop(['Geographical Coordinates', 
                                              'Unnamed: 0',
                                              'GDP (Adjusted to $US)', 
                                              'Real GDP (Purchasing Power Parity) 2020',
                                              'Real GDP Per Capita 2020',
                                              'Waste Generated Annually', 
                                              'Waste Recycled Annually',
                                              '% of Waste Recycled Annually',
                                              'Particulate Matter Emissions',
                                              'C02 Emissions', 
                                              'Methane Emissions',
                                              'Environmental Agreements (party to)',
                                              'Environmental Agreements (signed but not ratified)',                                             
                                              'Electricity by Fossil Fuels', 
                                              'Electricity by Nuclear',
                                              'Electricity by Solar', 
                                              'Electricity by Wind',
                                              'Electricity by Hydroelectricty', 
                                              'Electricity by Tide and Wave',
                                              'Electricity by Geothermal', 
                                              'Electricity by Biomass and Waste',
                                              'Urban Population (%)', 
                                              'Net Migration',
                                              'Environment Agreements #',                                              
                                              'Environment Agreements Not Ratified #'
                                             ],axis=1)

# Add in cleaned columns
final_country_data_df['Lat']=lats
final_country_data_df['Long']=lons
final_country_data_df['GDP (billion USD)']=GDP_Adjusted_to_USD
final_country_data_df['Real GDP 2020 (billion USD)']=Real_GDP
final_country_data_df['Real GDP Per Capita 2020']=GDP_per_Capita
final_country_data_df['Wasted Generated Annually (million tons)']=Waste_generated_annually
final_country_data_df['Wasted Recycled Annually (million tons)']=Waste_recycled_annually
final_country_data_df['% of Wasted Recycled Annually']=Percent_of_Waste_Recycled_Annually
final_country_data_df['Net Waste']=dev_netwaste
final_country_data_df['Particulate Matter Emissions (micrograms per m3)']=Particulate_Matter_Emissions
final_country_data_df['C02 Emissions (megatons)']=C02_Emmissions
final_country_data_df['Methane  Emissions (megatons)']=Methane_Emissions
final_country_data_df['Electricity by Unclean Energy (%)']=Electricity_by_Fossil_Fules
final_country_data_df['Electricity by Clean Energy (%)']=[100-row for row in Electricity_by_Fossil_Fules]
final_country_data_df['Urban Population (%)']=Urban_Population
final_country_data_df['Net Migration']=Net_Migration
final_country_data_df['Total Agreement']=Net_Migration
final_country_data_df['Development Status']=country_status

# Display DataFrame
final_country_data_df.head()
```

```python
# Export the City_Data into a csv
final_country_data_df.to_csv("../output_data/Cleaned_Countries_Data.csv")
```
